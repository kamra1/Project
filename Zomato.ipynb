```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.location = le.fit_transform(df.location)
df.rest_type = le.fit_transform(df.rest_type)
df.cuisines = le.fit_transform(df.cuisines)
```

```python
df_scaled = df.copy()
col_names = ['rate', 'votes','avg_cost']
features = df_scaled[col_names]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(features.values)
df_scaled[col_names]= sc.transform(features.values)
```

```python
df_scaled
```

```python
x = df_scaled.drop(['name','dish_liked','listed_type','rate'],axis=1)
y=df_scaled.rate
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.25,random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred_lr = lr.predict(x_test)
```

```python
lr.score(x_test, y_test)
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred_lr))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred_lr))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred_lr)))
```

# All of our errors are high in linear regression model

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=40, random_state = 40)
rf.fit(x_train,y_train)
```

```python
y_pred_rf = rf.predict(x_test)
```

```python
r2_score(y_test,y_pred_rf)
```
